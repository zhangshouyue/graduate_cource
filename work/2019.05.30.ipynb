```python
a =hdu['pmRA_x']
b =hdu['pmDE']
c =hdu['Jmag']
d =hdu['Kmag']
y = d + np.log((a**2+b**2)**0.5) -10
x = c-d
a1 =hdu1['pmRA']
b1 =hdu1['pmDE']
c1 =hdu1['Jmag']
d1 =hdu1['Kmag']
y1 = d1 + np.log((a1**2+b1**2)**0.5) -10
x1 = c1-d1
```

```python
plt.scatter(x,y,c='aqua',label='SAMPLE',s=1)
plt.scatter(x1,y1,c='r',label='OB',s=1)
plt.title('The reduced proper motion',color='y',fontsize=12)
plt.xlabel('J-K',color='crimson',fontsize=10)
plt.ylabel('H_nu',color='crimson',fontsize=10)
plt.legend(loc='upper right')
#plt.show()
plt.savefig('/home/zkc/桌面/90星图/_06.21.01.png',dpi=100)
```
