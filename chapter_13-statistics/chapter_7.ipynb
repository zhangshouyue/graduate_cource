```python
from scipy import interpolate
from numpy import polynomial as P
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
p1 = P.Polynomial([1, 2, 3])
```

```python
p1
```

```python
np.__version__#两个下划线
```

```python
p2 = P.Polynomial.fromroots([-1, 1])
```

```python
p2
```

```python
p1.roots()
```

```python
p2.roots()
```

```python
fig=plt.figure(figsize=(10,8))
x=np.linspace(-5,5)
plt.plot(x,p1(x),label=str(p1))
plt.plot(x,p2(x),label=str(p2))
plt.axhline(0,ls=':')
plt.legend()
ax_insert=fig.add_axes([0.35,0.4,0.35,0.45])
ax_insert.plot(x,p1(x))
ax_insert.plot(x,p2(x))
ax_insert.set_xlim([-2,2])
ax_insert.set_ylim([-2,6])
ax_insert.axhline(0,ls=':')
```
